# 08 - RAG Benchmark Suite (Retrieval + Generation + End-to-End)

This notebook runs a structured benchmark on all SRS files using question sets under:
`data/benchmarks/rag_eval/questions/`.

Outputs are saved under:
`data/outputs/evaluation/rag_benchmark/`

```python
from pathlib import Path
import subprocess
import sys
import pandas as pd
```

```python
RAG_LAB = Path("..").resolve()
SCRIPT = RAG_LAB / "scripts" / "run_rag_benchmark.py"

cmd = [
    sys.executable,
    str(SCRIPT),
    "--max-questions-per-srs",
    "10",
    "--k",
    "5",
    "--threshold",
    "0.72",
    "--answer-temperature",
    "0.1",
]

proc = subprocess.run(cmd, cwd=RAG_LAB, text=True, capture_output=True)
print(proc.stdout)
if proc.returncode != 0:
    print(proc.stderr)
    raise RuntimeError(f"Benchmark failed with code {proc.returncode}")
```

```python
OUT = RAG_LAB / "data" / "outputs" / "evaluation" / "rag_benchmark"

overall = pd.read_csv(OUT / "overall_summary.csv")
retrieval = pd.read_csv(OUT / "retrieval" / "retrieval_summary_by_srs.csv")
generation = pd.read_csv(OUT / "generation" / "generation_summary_by_srs.csv")
e2e = pd.read_csv(OUT / "e2e" / "e2e_summary_by_srs.csv")
failures = pd.read_csv(OUT / "diagnostics" / "failure_breakdown.csv")
by_category = pd.read_csv(OUT / "diagnostics" / "metrics_by_category.csv")

print("Overall")
display(overall)
print("Retrieval by SRS")
display(retrieval)
print("Generation by SRS")
display(generation)
print("End-to-end by SRS")
display(e2e)
print("Failure breakdown")
display(failures)
print("Metrics by category")
display(by_category)
```
